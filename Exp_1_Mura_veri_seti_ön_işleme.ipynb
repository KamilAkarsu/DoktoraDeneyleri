```python
import os
import cv2
import zipfile
import shutil
from google.colab import drive
from tqdm import tqdm

# 1. Google Drive'ı Colab ortamına bağlama
drive.mount('/content/drive')

# 2. Dosya Yolları (Bu kısımları kendi Drive dizininize göre güncelleyiniz)
# Örnek olarak MURA veri seti kullanılmıştır.
HAM_ZIP_YOLU = '/content/drive/MyDrive/Doktora/Verisetleri_tik/MURA_3.3GB.zip'
ISLENMIS_ZIP_YOLU = '/content/drive/MyDrive/Doktora/Verisetleri_tik/Islenmis_MURA.zip'

# Colab yerel çalışma dizinleri
YEREL_HAM_KLASOR = '/content/ham_veri'
YEREL_ISLENMIS_KLASOR = '/content/islenmis_veri'

HEDEF_BOYUT = (224, 224) # Hem geleneksel CNN'ler hem de standart ViT için optimal başlangıç boyutu

# 3. Gerekli klasörleri oluşturma
os.makedirs(YEREL_HAM_KLASOR, exist_ok=True)
os.makedirs(YEREL_ISLENMIS_KLASOR, exist_ok=True)

# 4. Ham Zip dosyasını Colab yerel diskine çıkartma
print("Zip dosyası çıkartılıyor...")
with zipfile.ZipFile(HAM_ZIP_YOLU, 'r') as zip_ref:
    zip_ref.extractall(YEREL_HAM_KLASOR)

# CLAHE (Contrast Limited Adaptive Histogram Equalization) Objesi
# ClipLimit ve tileGridSize değerleri radyolojik görüntüler için literatürde sıkça tercih edilen standart değerlerdir.
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# 5. Görüntüleri işleme döngüsü
print("Görüntüler işleniyor (Yeniden boyutlandırma ve CLAHE)...")
desteklenen_formatlar = ('.png', '.jpg', '.jpeg', '.dcm')

for root, dirs, files in os.walk(YEREL_HAM_KLASOR):
    for file in tqdm(files):
        if file.lower().endswith(desteklenen_formatlar):
            dosya_yolu = os.path.join(root, file)

            # Görüntüyü gri tonlamalı (grayscale) olarak okuma
            img = cv2.imread(dosya_yolu, cv2.IMREAD_GRAYSCALE)

            if img is not None:
                # a. Yeniden Boyutlandırma
                img_resized = cv2.resize(img, HEDEF_BOYUT, interpolation=cv2.INTER_CUBIC)

                # b. CLAHE Uygulaması (Kontrast artırımı ile kırık hatlarının belirginleştirilmesi)
                img_clahe = clahe.apply(img_resized)

                # İşlenmiş veriyi kaydetmek için orijinal klasör yapısını koruma
                bagil_yol = os.path.relpath(root, YEREL_HAM_KLASOR)
                hedef_klasor_yolu = os.path.join(YEREL_ISLENMIS_KLASOR, bagil_yol)
                os.makedirs(hedef_klasor_yolu, exist_ok=True)

                hedef_dosya_yolu = os.path.join(hedef_klasor_yolu, file)
                cv2.imwrite(hedef_dosya_yolu, img_clahe)

# 6. İşlenmiş verileri yeni bir Zip dosyası haline getirme
print("İşlenmiş veriler sıkıştırılıyor...")
shutil.make_archive('/content/Islenmis_MURA', 'zip', YEREL_ISLENMIS_KLASOR)

# 7. Oluşturulan yeni Zip dosyasını kalıcı olarak Google Drive'a taşıma
print("İşlenmiş veri seti Google Drive'a kopyalanıyor...")
shutil.copy('/content/Islenmis_MURA.zip', ISLENMIS_ZIP_YOLU)

# 8. Colab yerel diskini temizleme (RAM/Disk optimizasyonu için)
shutil.rmtree(YEREL_HAM_KLASOR)
shutil.rmtree(YEREL_ISLENMIS_KLASOR)
os.remove('/content/Islenmis_MURA.zip')

print(f"İşlem başarıyla tamamlandı. İşlenmiş veri seti şuraya kaydedildi: {ISLENMIS_ZIP_YOLU}")
```
